The for loop will use the .walk function to recursively print the directories.'dirname' represents the the outer level and file names represents the inner level. '_' is meant as a placeholder as it will not be used in the code. A nested loop is created to iterate throught the inner file names. Within the nested loop, os.path.join is used to join the outer and inner components. Each join is printed to screen, displaying the system path for later access.

```python
train_data = pd.read_csv("/kaggle/input/titanic/train.csv")
train_data.head()
```

```python
test_data = pd.read_csv("/kaggle/input/titanic/test.csv")
test_data.head()
```

```python
# Using label-based selection (.loc) to identify females within the training data.
# Return the indicator of survival (0 or 1) for those respective values.
# Print the percentage of men and women among those who survived.

women = train_data.loc[train_data.Sex == 'female']["Survived"]

rate_women = round(sum(women)/len(women) * 100,2)
print("Women who survived:", str(rate_women) + "%")

men = train_data.loc[train_data.Sex == 'male']["Survived"]

rate_men = round(sum(men)/len(men) * 100,2)
print("Men who survived:", str(rate_men) + "%")

##################################################################################

# Using a for loop to sum all the passengers and those who survived.
# Calculating what percentage of the passengers survived.

total_survival = train_data ["Survived"]

count_survive = 0
count_total = 0

for num in total_survival:
    count_total += 1
    if num == 1:
        count_survive += 1
        
survive_percent = str(round((count_survive / count_total * 100), 2)) + "%"
print("Passengers who survived:", survive_percent)
```

```python
from sklearn.ensemble import RandomForestClassifier

y = train_data["Survived"]

features = ["Pclass", "Sex", "SibSp", "Parch"]
X = pd.get_dummies(train_data[features])
X_test = pd.get_dummies(test_data[features])

model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X, y)
predictions = model.predict(X_test)

output = pd.DataFrame({'PassengerId': test_data.PassengerId, 'Survived': predictions})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```
